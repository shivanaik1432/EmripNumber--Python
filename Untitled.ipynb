# Emrip Number


# An Emirp Number (prime spelled backwards) is a prime number that results in a different prime when its decimal digits are reversed. This definition excludes the related palindromic primes.

```python
# Checking given number is emrip number or not

number=int(input("Enter ur Checked number: "))
dummy=number
total=0
while dummy>0:
    rem=dummy%10
    total=total*10+rem
    dummy//=10
if total!=number:
    if number>1:
        for i in range(2,number//2+1):
            if number%i==0:
                print(f'{number} is not Emrip number')
                break
        else:
            for j in range(2,total//2+1):
                if total%j==0:
                    print(f'{number} is not a Emrip number')
                    break
            else:
                print(f'{number} is a Emrip number')
    else:
         print(f'{number} is not a Emrip number')
else:
     print(f'{number} is not a Emrip number')
```
